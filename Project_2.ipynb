```python
# Беремо колонку з мовами
languages = survey["LanguageHaveWorkedWith"]

# Перевіряємо, чи є Python у рядку (True для тих, у кого є Python) 
python_lang = languages.str.contains("Python", na=False,case=False) 
# case=False захистить, якщо Pyton буде з маленької літери

# Обчислюємо відсоток
python_percent = python_lang.mean() * 100  # mean() працює, бо True=1, False=0
# python_lang виглядає, як [True, False, True, True, False] → [1, 0, 1, 1, 0]
# .mean() рахує середнє значення:
# Наприклад (1 + 0 + 1 + 1 + 0) / 5 = 3 / 5 = 0.6

# Округлюємо до 1 знаку після коми
python_percent = round(python_percent, 1)
print("Відсоток респондентів, які програмують на Python:"f"{python_percent}%")
```

```python
# Завдання 7. Географічний аналіз компенсації Python-розробників

#  Фільтруємо Python-розробників
python_dev = survey[survey["LanguageHaveWorkedWith"].str.contains("Python", na=False,case=False)]
# Спочатку створюється список з True/False, потім відбираємо в список тільки з True, тому survey[survey[
print(f"ConvertedCompYearly до фільтрації: {len(python_dev):} ")

# Видаляємо пропуски в стовпчику компенсації ConvertedCompYearly
python_dev = python_dev.dropna(subset=["ConvertedCompYearly"])
print(f"ConvertedCompYearly після фільтрації: {len(python_dev):} ")

# Групуємо по країнах і рахуємо метрики
result = python_dev.groupby("Country")["ConvertedCompYearly"].agg(
    mean_comp= "mean",
    median_comp="median"
).round(0).reset_index()

result
```

```python
# Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів

# Перевіряємо ResponseId на дублікати на нулі
resp_duplicates= survey["ResponseId"].duplicated()
print(f"Total duplicate rows: {resp_duplicates.sum():,}")
resp_null= survey["ResponseId"].isnull()
print(f"Total null rows: {resp_null.sum():,}")

# Видаляємо пропуски
survey_comp = survey.dropna(subset=["ConvertedCompYearly"])
# Сортуємо по зарплаті (від більшої до меншої)
top5 = survey_comp.sort_values(by="ConvertedCompYearly", ascending=False).head(5)
# Дивимось рівень освіти
top5[["ConvertedCompYearly", "EdLevel","Country"]] # Country - просто цікаво по країнам
```

```python
# Мені цікаво: Я бачу якісь дуже великі суми, можливо аномальні викиди
# Тому прибрала екстремальні значення
#  Беремо колонку ConvertedCompYearly
comp = survey["ConvertedCompYearly"].dropna()
# Рахуємо вищу межу
high = comp.quantile(0.95)
# 3. Фільтруємо
filtered = survey[
        (survey["ConvertedCompYearly"] <= high)
]
# Відфільтрую лише унікальні значення зарплати, тому що пробувала без фільтра, видає однакові значачення
top_5_quant = filtered.drop_duplicates(subset=["ConvertedCompYearly"])\
                      .sort_values(by="ConvertedCompYearly", ascending=False)\
                      .head(10)
# Виводимо потрібні колонки
top_5_quant[["ConvertedCompYearly", "EdLevel","Country"]]
# Можливо це більш реалістичні топ 5 компенсацій
```

```python
# Цікаво подивитися розподіл виплат
filtered["ConvertedCompYearly"].value_counts().head(10)
```

```python
# Завдання 9. Аналіз популярності Python по віковим категоріям 

# Подивимось колонку 'Age' по віковим категоріям
survey['Age'].value_counts()
```

```python
# Перевіримо 
survey['Age'].isna().sum()
```

```python

# Створюємо нову булеву колонку 'python_lang', де True - це входить Python
survey['python_lang'] = survey['LanguageHaveWorkedWith'].str.contains("Python", na=False,case=False)
# survey['python_lang']
result = (
    survey
    .groupby('Age')['python_lang']
    .mean() # Рахуємо середнє значення (mean = (1 + 0 + 1) / 3 = 0.67)
    .mul(100) # Множимо на 100
    .reset_index()
    .rename(columns={'python_lang': 'python_percent'})
)
# Округлимо
result['python_percent'] = result['python_percent'].round(2)

print(result)
```

```python
# Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників
comp = survey["ConvertedCompYearly"].dropna()
# Рахуємо вищу межу
high = comp.quantile(0.75)
# 3. Фільтруємо
filtered = survey[
        (survey["ConvertedCompYearly"] >= high)&(survey["RemoteWork"] == "Remote")]

industry = filtered["Industry"].value_counts().head(10)

industry = industry.reset_index()
industry.columns = ["Industry", "Count"]
# Дивимось 
print(industry)
```
